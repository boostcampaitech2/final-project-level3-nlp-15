```python
import os
ROOT_PATH = os.path.abspath(".") # this makes compatible absolute path both for local and server
DATA_PATH = os.path.join(ROOT_PATH, "_clones", "dataset_clones")
POKEDEX = os.path.join(DATA_PATH,"pokedex")
```

```python
import os
import pandas as pd
import numpy as np

characters_csv = os.path.join(DATA_PATH, "All_Pokemon.csv")
description_csv = os.path.join(DATA_PATH, "Finally_Fully_Scraped_Pokemon_Data.csv")
```

```python
df_characters = pd.read_csv(characters_csv)
df_description = pd.read_csv(description_csv)
```

```python
# pandas display all columns
pd.set_option('display.max_columns', None)

# pandas display full length values
pd.set_option('display.max_colwidth', None)

display(df_characters.head())
display(df_description.head())
```

```python
print(df_characters.columns)
print(df_description.columns)
```

```python
columns_1 = \
[
       'Number', 'Name', 'Type 1', 'Type 2', "Height", "Weight", "Generation", "BMI" # BMI can later converted into size (small, medium, large)
]

df_char_filtered = df_characters[columns_1]

# rename df_char_filtered's colunm "Name" as "pokemon_name"
df_char_filtered = df_char_filtered.rename(
       columns={
              "Name": "pokemon_name",
              "Type 1": "type_main",
              "Type 2": "type_sub",
              }
       )


columns_2 = \
[
       'pokemon', 'Description_X', 'Description_Y', 'Category', 'Ability_1', 'Ability_2', \
       'Weaknesses', 'HP', 'Attack', 'Defense', 'Sp.Atk', 'Sp.Def', 'Speed','Total', \
       'Evolutions', 'Evolution_type', 'Hidden_Ability', 'evolution_list_index'
]

df_desc_filtered = df_description[columns_2]

# rename column 
df_desc_filtered = df_desc_filtered.rename(
       columns={
              "pokemon": "pokemon_name",
              "Description_X": "description_main",
              "Description_Y": "description_sub",
              "Ability_1": "ability_main",
              "Ability_2": "ability_sub",
              "Type 1": "type_main",
              "Type 2": "type_sub",
              }
       )

display(df_char_filtered.head())
display(df_desc_filtered.head())
```

```python
df_merged = pd.merge(df_char_filtered, df_desc_filtered, on="pokemon_name")
# lowercase all column names from df_merged
df_merged.columns = [x.lower() for x in df_merged.columns]
df_merged.head()
```

```python
df_merged.to_csv(os.path.join(ROOT_PATH, "pokemon_data.csv"), index=False)
```
